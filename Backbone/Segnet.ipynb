# Segnet

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten , Dense , MaxPool2D  , Conv2D  , UpSampling2D , BatchNormalization
```

segnet.png

```python
segnet = Sequential([
    Conv2D(64 , (3 , 3) , padding = 'same' , activation = 'relu' , input_shape = (256 , 256 , 1)) ,
    BatchNormalization(),
    MaxPool2D(),
    Conv2D(128 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    MaxPool2D(),
    Conv2D(256 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    MaxPool2D(),
    Conv2D(512 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    MaxPool2D(),

    Conv2D(512 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    UpSampling2D(),
    Conv2D(256 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    UpSampling2D(),
    Conv2D(128 , (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    UpSampling2D(),
    Conv2D(64, (3 , 3) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
    UpSampling2D(),

    Conv2D(1 , (1 , 1) , padding = 'same' , activation = 'relu') ,
    BatchNormalization(),
])
```

```python
segnet.compile(loss = 'mse' , optimizer = 'adam')
```
