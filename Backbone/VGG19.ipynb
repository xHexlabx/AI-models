```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten , Dense , MaxPool2D  , Conv2D  
```

llustration-of-the-network-architecture-of-VGG-19-model-conv-means-convolution-FC-means.png

```python
vgg19 = Sequential([

    Conv2D(64  , (3 , 3) , padding = 'same' , activation = 'relu' , input_shape = (224 , 224 , 3)) ,
    Conv2D(64  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    MaxPool2D() ,

    Conv2D(128  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(128  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    MaxPool2D() ,

    Conv2D(256  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(256  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(256  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(256  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    MaxPool2D() ,

    Conv2D(512 , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    MaxPool2D() ,

    Conv2D(512 , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    Conv2D(512  , (3 , 3) , padding = 'same' , activation = 'relu' ) ,
    MaxPool2D() ,

    Flatten() ,
    Dense(4096 , activation = 'relu') ,
    Dense(4096 , activation = 'relu') ,
    Dense(1000 , activation = 'softmax')
])
```

```python
vgg19.compile(loss = 'mse' , optimizer = 'adam' , metrics = 'accuracy')
```
